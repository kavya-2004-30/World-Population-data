```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/kaggle/input/world-data-population/world_population_data.csv")
plt.figure(figsize=(9, 8))
sns.barplot(x=df['continent'], y=df['2023 population'], color='gray')
ax =sns.barplot(x=df['continent'], y=df['1970 population'],color='pink').set_ylabel('population of continents', fontsize=10)
plt.title('Population change over the entire period');
```

```python
sizes=[24.36,21.37,21.37,17.09,9.83,5.98]
labels=['Africa','Asia','Europe','North America','Oceania','South America']
plt.style.use('classic')
plt.title('Population of World')
plt.pie(sizes,labels=labels,autopct='%1.2f%%',labeldistance=1.1,startangle=90,shadow=True,explode=[0.2,0,0,0,0,0.2])
plt.show()
```

```python
topfive=df.head(5)
topfive_pop = topfive["2023 population"]
topfive_country=topfive["country"]
plt.figure(figsize=(20,6),dpi=300)
sns.barplot(data=topfive,x=topfive_country,y=topfive_pop)
```

```python
topten=df.head(10)
topten_pop = topten["2023 population"]
topten_country=topten["country"]
plt.figure(figsize=(20,6),dpi=300)
sns.barplot(data=topten,x=topten_country,y=topten_pop)
```

```python
d=df.iloc[:,3:13]
per=d.columns[1:10].values[::1].tolist()
a= {pr : pr.replace(" population", "") for pr in per}
pop_growth=d.groupby('continent')[d.columns[1:10].values[::1]].sum().T
pop_growth = pop_growth.rename(index=a)
pop_growth
```

```python
sns.heatmap(pop_growth);
```

```python
import matplotlib.pyplot as plt
colors = {
    'Asia': 'red',
    'Europe': 'blue',
    'Africa': 'green',
    'North America': 'orange',
    'South America': 'purple',
    'Oceania': 'brown',
    'Antarctica': 'gray'
}
ax = pop_growth.plot(figsize=(12, 7), linewidth=3, color=[colors.get(col, 'black') for col in pop_growth.columns])
plt.title('Population growth by Continent (2014–2023)', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend(title='Continent')
plt.tight_layout()
plt.show()
```
